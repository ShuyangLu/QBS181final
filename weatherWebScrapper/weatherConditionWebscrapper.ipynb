```python
# def
def ks(url,tmoo):
    driver.get(url)
    kaisw = 0
    for II in range(10):
        sj = driver.page_source
        # print(sj)
        sj = etree.HTML(sj)
        li_s = sj.xpath('//div[@class="summary-table"]/table/tbody[1]/tr/td[1]/text()')
        # print(li_s)
        if len(li_s) == 0:
            time.sleep(5)
            continue
        else:
            kaisw = 1
            li_s2 = sj.xpath('//div[@class="summary-table"]/table/tbody[2]/tr/td[1]/text()')
            li_s3 = sj.xpath('//div[@class="summary-table"]/table/tbody[3]/tr[4]/td[1]/text()')

            sd2 = sj.xpath('//tbody[@role="rowgroup"]/tr/td[4]/lib-display-unit/span/span/text()')
            print('-------------')
            print(li_s2)
            print(li_s3)
            print(sd2)
            sd = []
            for i in sd2:
                sd.append(int(i))
            if len(sd) == 0:
                sd = ''
            else:
                sd = sum(sd)/len(sd)

            if len(li_s2) == 0:
                li_s2 = ['']
            if len(li_s3) == 0:
                li_s3 = ['']
            if len(li_s) != 3:

                li_s2s = ['','','']
            else:
                li_s2s = []
                for i in li_s:
                    li_s2s.append(str(i))




            break
    if kaisw == 1:
        # write down into a csv
        print(','.join(li_s2s)+','+str(li_s2[0])+','+str(li_s3[0])+','+str(sd)+'\n')
        ss = open('sj.csv','a',encoding='gbk')
        ss.write(','.join(li_s2s)+','+str(li_s2[0])+','+str(li_s3[0])+','+str(sd)+','+str(tmoo)+','+'new york city'+'\n')
        ss.close()
    # input()





```
